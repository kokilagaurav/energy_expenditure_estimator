```python
# Remove height, weight, and heart rate outliers
df_cleaned = df[
    (df['Height'] <= 220) & (df['Height'] >= 125) &
    (df['Weight'] <= 120) & (df['Weight'] >= 40) &
    (df['Heart_Rate'] <= 120) & (df['Heart_Rate'] >= 50)
]
df_cleaned.head()
df_cleaned.shape
```

```python
skew_cols = []

num_cols = df.select_dtypes(include=[np.number]).columns

for col in num_cols:

    if df[col].skew() > 0.5 or df[col].skew() < -0.5:

        skew_cols.append(col)

        print(f"{col} is skewed with a skewness of {df[col].skew()}")

    else:

        print(f"{col} is not skewed ")
```

```python
from scipy.stats import boxcox

for col in skew_cols:
    # Ensure all values are positive for Box-Cox transformation
    if (df[col] <= 0).any():
        df[col] = df[col] + abs(df[col].min()) + 1
    
    df[col] , _ = boxcox(df[col])
    if df[col].skew() > 0.5 or df[col].skew() < -0.5:
        print(f"{col} is skewed with a skewness of {df[col].skew()}")
    else:
        print(f"{col} is not skewed ")
```
